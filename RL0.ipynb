```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Données d'exemple
surface = np.array([50, 60, 70, 80, 90, 100]).reshape(-1, 1)
prix = np.array([150000, 180000, 210000, 240000, 270000, 300000])

# Modèle de régression linéaire
model = LinearRegression()
model.fit(surface, prix)

# Prédictions
prix_pred = model.predict(surface)

# Visualiser les résultats
plt.scatter(surface, prix, color='blue', label='Données réelles')
plt.plot(surface, prix_pred, color='red', label='Régression linéaire')
plt.xlabel('Surface (m²)')
plt.ylabel('Prix (€)')
plt.legend()
plt.show()
```
